## Monte Carlo Analysis

```python
# Setting up the time horizon
days = 365
dt = 1/days

#  mu (drift) from the expected return data for OXY
mu = risk_an.mean()['OXY']

# the volatility of the stock from the std() of the average return
sigma = risk_an.std()['OXY']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    # takes in starting stock price, days of simulation,mu,sigma
    #and returns simulated price array
    
    # Define a price array
    price = np.zeros(days)
    price[0] = start_price
    # shock and drift
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    # Run price array for number of days
    for x in range(1,days):
        
        # Calculate Schock
        shock[x] = np.random.normal(loc=mu * dt, scale=sigma * np.sqrt(dt))
        # Calculate Drift
        drift[x] = mu * dt
        # Calculate Price(the previous +p)
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
```

```python
OXY.head()
```

```python
# Get start price from OXY.head()
start_price =83.04

for run in range(100):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
plt.xlabel("Days")
plt.ylabel("Price")  
plt.title('Monte Carlo Analysis for OXY');
```

```python
# Setting a large numebr of runs
runs = 10000

# Creating an empty matrix to hold the end price data
simulations = np.zeros(runs)

# Setting the print options of numpy to only display 0-5 points from an array to suppress output
np.set_printoptions(threshold=5)

for run in range(runs):    
    # Setting the simulation data point as the last stock price for that run
    simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1];
```

```python
#  defining q as the 1% empirical qunatile, this basically means that 99% of the values should fall between here
q = np.percentile(simulations, 1)
    
#  plotting the distribution of the end prices
plt.hist(simulations,bins=200)

# Using plt.figtext to fill in some additional information onto the plot

# Starting Price
plt.figtext(0.6, 0.8, s="Start price: $%.2f" %start_price)
# Mean ending price
plt.figtext(0.6, 0.7, "Mean final price: $%.2f" % simulations.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,))

# Displaying 1% quantile
plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q)

# Plotting a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# Title
plt.title(u"Final price distribution for OXY Stock after %s days" % days, weight='bold');
```

##  Logistic Regression

```python
MRO = MRO.dropna()
MRO= MRO.iloc[:,:4]
MRO.head()
```

```python
#MRO['Daily Return']= df.dropna()
X = MRO.iloc[:,:9]
```

```python
y = np.where(MRO['Close'].shift(-1) > MRO['Close'],1,-1)
```

```python
X.shape,y.shape
```

```python
#Spliting the Dataset
split = int(0.7*len(MRO))
X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
lr = LogisticRegression()
lr=LogisticRegression().fit(X_train,y_train)
```

```python
#looking at the coefficients
pd.DataFrame(zip(X.columns, np.transpose(lr.coef_)))
```

```python
#calculating the probabilities 
probability =  lr.predict(X_test)
prediction = lr.predict(X_test)
print (probability)
```

```python
print(metrics.classification_report(y_test, prediction))
```

```python
print(lr.score(X_test,y_test))
print(lr.score(X,y))
```

```python
c_val = cross_val_score(lr, X, y, cv=10)
print(c_val)
```

```python
print(c_val.mean())
```
